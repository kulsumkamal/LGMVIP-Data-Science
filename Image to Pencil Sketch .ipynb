# Image to pencil sketch using Python 

The image read in RBG format, then converted into a grayscale image. The grayscale image is then inverted to obtain a negative image ehich is done in order to enhance details. Finally, the pencil sketch is created by mixing the grayscale image with the inverted blurry image.

```python
import cv2
from matplotlib import pyplot as plt
```

Reading the image.

```python
image = cv2.imread('C:/Users/kulsum kamal/Downloads/horse.jpg')
image.shape
```

```python
# Matplotlib intepretes images in RGB format whereas cv2 uses BGR format
# So in order to display it we must first convert it 
color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (12,12))
plt.imshow(color)
plt.title('Horse')
plt.axis('off')
plt.show()
```

Converting to gray scale.

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize = (12,12))
plt.imshow(gray, cmap= 'gray')
plt.title('Grayscale image')
plt.axis('off')
plt.show()
```

inverting the grayscale image.

```python
inv = cv2.bitwise_not(gray)

plt.figure(figsize = (12,12))
plt.imshow(inv, cmap = 'gray')
plt.title('Inverted image')
plt.axis('off')
plt.show()
```

Blurring the image using gaussian function.

```python
blurred = cv2.GaussianBlur(inv, (21, 21), 0)

plt.figure(figsize = (12,12))
plt.imshow(blurred, cmap='gray')
plt.title('Smoothened image')
plt.axis('off')
plt.show()
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray, inverted_blurred, scale=255)

plt.figure(figsize = (12,12))
plt.imshow(pencil_sketch, cmap = 'gray')
plt.title('Pencil sketch')
plt.axis('off')
plt.show()
```
